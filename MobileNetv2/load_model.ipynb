```python
import tensorflow as tf
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
def normalize_and_resize(img_loc):
    # With Google Colab
    # my_img = tf.io.read_file(img_loc)

    # Read in image
    my_img = tf.io.read_file(img_loc)

    # Turn file into a tensor
    my_img = tf.image.decode_image(my_img)

    # Resize image
    my_img = tf.image.resize(my_img, size=[224, 224])

    # Normalize data
    my_img = my_img / 255

    return my_img
# Load the saved model
model_path ="../MobileNetv2/mobilenet_smartbin"
loaded_model = tf.keras.models.load_model(model_path)
# Make predictions on new data
class_names = ['Inorganic','Organic','Metal']
image_path=  "/Users/maniksinghsarmaal/Downloads/burger.jpeg"
custom_image = normalize_and_resize(image_path)
img = Image.open(image_path)
plt.imshow(img)
plt.show()
prediction = loaded_model.predict(tf.expand_dims(custom_image, axis=0))
print(f'The class of the feed input image is : {class_names[np.argmax(prediction)]}')
```
